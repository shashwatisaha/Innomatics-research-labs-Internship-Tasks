Problem 1: Unique Words in a Sentence
Problem Statement
Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves


```python
# Input
sentence = input("Enter sentence: ")

# Split into words and convert to set for unique words
words = sentence.split()
unique_words = set(words)

# Display results
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

Problem 2: Highest Salary from Employee Data
Problem Statement
Store employee names and their salaries, and determine which employee has the highest salary.


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000,
    "Shashwati": 90000
}

# Find employee with highest salary using max()
max_employee = max(employees, key=employees.get)
highest_salary = employees[max_employee]

print(f"Highest Salary: {max_employee} - {highest_salary}")
```

Problem 3: Find Maximum and Minimum Values
Problem Statement
Find the maximum and minimum values from a list of numbers.


```python
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through list to find max and min
for num in numbers[1:]:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
    else:
        pass

print(f"List: {numbers}")
print(f"Max: {maximum}")
print(f"Min: {minimum}")
```

Problem 4: Count Products Above a Price Threshold
Problem Statement
Count how many products have a price greater than 1000.


```python
prices = [450, 1200, 899, 1500, 300,2000]

# Initialize counter
count = 0

# Loop through prices and count those above 1000
for price in prices:
    if price > 1000:
        count += 1

print(f"Products above 1000: {count}")
```

Problem 5: Calculate Attendance Percentage
Problem Statement
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
attendance = ["P", "P", "A", "P", "P"]

# Count present days
present_days = 0
for day in attendance:
    if day == "P":
        present_days += 1

# Calculate percentage
total_days = len(attendance)
percentage = (present_days / total_days) * 100

print(f"Attendance Percentage: {percentage}")
```

Problem 7: Count Character Frequency
Problem Statement
Count how many times each character appears in a given string.


```python
text = "Python Programming"

# Initialize empty dictionary for character counts
char_count = {}

# Count frequency of each character
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

Problem 10: Calculate Average Marks
Problem Statement
Calculate the average marks of a student from a list of marks.


```python
marks = [60, 85, 80, 70, 90]

# Calculate average using sum and length
total = sum(marks)
average = total / len(marks)

print(f"Average Marks: {average}")
```
