Problem Statement 1: Employee Performance Bonus Eligibility


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = max(employees.values())  # Finds 92 [code_file:1]
top_performers = [name for name, score in employees.items() if score == highest_score]  # Ravi, Kiran [code_file:1]

output = f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})"
print(output)  # Matches expected: Top Performers Eligible for Bonus: Ravi, Kiran (Score: 92) [code_file:1]
```

Problem Statement 2: Search Query Keyword Analysis

```python
import string
from collections import Counter

query = "Buy mobile phone buy phone online"

# Clean: lowercase and remove punctuation [code_file:1]
query_lower = query.lower()
translator = str.maketrans('', '', string.punctuation)
clean_query = query_lower.translate(translator)

# Count and filter frequencies > 1 [code_file:1]
words = clean_query.split()
freq = Counter(words)
result = {k: v for k, v in freq.items() if v > 1}

print(result)  # {'buy': 2, 'phone': 2} [code_file:1]
```

Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Filter even values with enumerate for (index, value) [code_file:1]
valid_readings = [(i, val) for i, val in enumerate(sensor_readings) if val % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)  # [(1, 4), (3, 8), (4, 10), (5, 12)] [code_file:1]
```

Problem Statement 4: Email Domain Usage Analysis


```python
from collections import Counter

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Extract domains and count [code_file:1]
domains = [email.split('@')[1] for email in emails]
counts = Counter(domains)
total = len(emails)

# Print percentages rounded to nearest whole number [code_file:1]
for domain, count in counts.items():
    percentage = round((count / total) * 100)
    print(f"{domain}: {percentage}%")
```

Problem Statement 5: Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average: (1200+1500+900+2200+1400+3000)/6 = 1600
average = sum(sales) / len(sales)

# Detect spikes: > 30% above average (threshold: 1600 * 1.3 = 2080)
for day, sale in enumerate(sales, 1):
    if sale > average * 1.3:
        print(f"Day {day}: {sale}")
```
