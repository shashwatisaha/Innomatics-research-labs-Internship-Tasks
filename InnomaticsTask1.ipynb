3. Simple Data Cleaner
Problem Statement:

4. Message Length Analyzer
Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]


```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(f"'{msg}' length: {length}", end="")
    if length > 10:
        print(" - Flag: Too long")
    else:
        print()
```
