Problem 1: Mobile Recharge Validation System
Create a function to validate a mobile recharge request.


```python
def validate_recharge():
    valid_plans = {199, 299, 399, 599}

    while True:
        try:
            amount = float(input("Enter recharge amount (₹): "))
            if amount < 50:
                print("Invalid: Amount must be ≥ ₹50. Please retry.")
                continue

            plan = int(input("Enter plan (199/299/399/599): "))
            if plan not in valid_plans:
                print(f"Invalid plan. Valid plans: {sorted(valid_plans)}. Please retry.")
                continue

            # Success: return values
            print(f"Recharge validated: ₹{amount} for plan {plan}.")
            return amount, plan

        except ValueError:
            print("Invalid input. Enter numbers only. Retry.")
```

```python
# Simulate usage
result = validate_recharge()
print("Processing recharge:", result)
```

Problem 2: Inventory Reorder Alert System
Given a dictionary of products and stock quantities:
-Print Reorder Alert if stock < 15
-Print Stock OK otherwise

```python
def check_inventory(inventory):
    """
    Checks stock levels and prints alerts.
    Args: inventory (dict) - {product: stock_qty}
    """
    for product, stock in inventory.items():
        if stock < 15:
            print(f"Reorder Alert: {product} (stock: {stock})")
        else:
            print(f"Stock OK: {product} (stock: {stock})")
```

```python
inventory = {
    "Laptop": 10,
    "Mouse": 20,
    "Keyboard": 5,
    "Monitor": 25
}
check_inventory(inventory)
```

Problem 3: Student Result Processing System
Create a function that:
-Accepts a list of student marks
-Calculates average marks
-Prints Pass if average ≥ 50 else Fail


```python
def process_student_result(marks):
    """
    Calculates average marks and determines pass/fail.
    Args: marks (list) - Student marks
    Returns: str - "Pass" or "Fail"
    """
    total = 0
    for mark in marks:      # Loop sums marks
        total += mark
    average = total / len(marks)
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"
    print(f"Average marks: {average:.2f} - Result: {result}")
    return result
```

```python
marks1 = [60, 70, 80]
result1 = process_student_result(marks1)  # Pass
print("Final result:", result1)
```

Problem 4: Cab Fare Estimator with Retry Option
Create a cab fare calculator:
-Base fare = ₹50
-₹12 per km
-Peak hour → 25% extra
-Allow retry using while loop

```python
def calculate_cab_fare():
    BASE_FARE = 50
    RATE_PER_KM = 12
    PEAK_SURCHARGE = 1.25

    while True:
        try:
            distance = float(input("Enter distance (km): "))
            if distance <= 0:
                print("Distance must be > 0. Retry.")
                continue

            peak_hour = input("Peak hour? (y/n): ").lower().startswith('y')

            fare = BASE_FARE + (distance * RATE_PER_KM)
            if peak_hour:
                fare *= PEAK_SURCHARGE

            print(f"Total fare: ₹{fare:.2f}")
            return fare

        except ValueError:
            print("Enter valid numbers. Retry.")
```

```python
fare = calculate_cab_fare()
```

Problem 5: Employee Attendance Eligibility Checker
Create a function that:
-Accepts attendance list ("P", "A")
-Calculates attendance percentage
-Returns Eligible if ≥ 75% else Not Eligible

```python
def check_attendance_eligibility(attendance):
    """
    Calculates attendance % and eligibility status.
    Args: attendance (list) - ["P", "A", "P", ...]
    Returns: str - "Eligible" or "Not Eligible"
    """
    present_days = 0
    total_days = len(attendance)

    for day in attendance:      # Loop counts "P"
        if day == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print(f"Attendance: {percentage:.1f}% ({present_days}/{total_days}) - {status}")
    return status
```

```python
attendance1 = ["P", "P", "A", "P", "P", "P"]
result = check_attendance_eligibility(attendance1)
print("Final status:", result)
```

Problem 6: Password Strength Checker
Create a function to check password strength.
Rules:
-Minimum length = 8
-Must contain at least one digit
-Must contain at least one special character (@#$)

```python
def check_password_strength(password):
    """
    Validates password against strength rules.
    Returns: str - "Strong" or specific failure reason
    """
    if len(password) < 8:
        return "Weak: Minimum length 8 required"

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for char in password:           # Loop checks each character
        if char.isdigit():
            has_digit = True
        elif char in special_chars:
            has_special = True

    if not has_digit:
        return "Weak: At least one digit required"
    if not has_special:
        return "Weak: At least one special char (@#$) required"

    return "Strong"
```

```python
passwords = ["abc123@xyz", "password123", "abc@def456", "weakpass"]
for pwd in passwords:
    result = check_password_strength(pwd)
    print(f"'{pwd}' -> {result}")
```
